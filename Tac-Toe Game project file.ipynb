# Tac-Toe Game

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Initialize the game state
board = [''] * 9  # Represents the Tic-Tac-Toe board
current_player = 'X'  # Start with player X
player1 = ''
player2 = ''

# Function to handle player names input and start the game
def start_game(_):
    global player1, player2
    player1 = player1_input.value.strip()
    player2 = player2_input.value.strip()
    if player1 and player2:
        display_game_interface()
    else:
        error_msg.value = "Please enter names for both players."

# Function to handle player move
def handle_move(button, index):
    global current_player
    
    if board[index] == '' and not check_winner() and not check_draw():
        board[index] = current_player
        button.description = current_player
        button.disabled = True
        button.style.font_weight = 'bold'  # Make the text bold
        button.style.font_size = '24px'  # Increase the font size
        if check_winner():
            winner_msg.value = f"Congratulations! {player1 if current_player == 'O' else player2} wins the game!"
            winner_msg.style = {'font_size': '24px', 'color': 'green', 'font_weight': 'bold'}
        elif check_draw():
            winner_msg.value = "No winner. Restart the game."
            winner_msg.style = {'font_size': '24px', 'color': 'red', 'font_weight': 'bold'}
        current_player = 'O' if current_player == 'X' else 'X'

# Function to check for a winner
def check_winner():
    winning_combinations = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),  # Horizontal
        (0, 3, 6), (1, 4, 7), (2, 5, 8),  # Vertical
        (0, 4, 8), (2, 4, 6)               # Diagonal
    ]
    
    for combo in winning_combinations:
        if board[combo[0]] == board[combo[1]] == board[combo[2]] != '':
            highlight_winning_line(combo)
            return True
    
    return False

# Function to highlight winning line
def highlight_winning_line(line):
    buttons[line[0]].style.button_color = 'lightgreen'
    buttons[line[1]].style.button_color = 'lightgreen'
    buttons[line[2]].style.button_color = 'lightgreen'

# Function to check for a draw
def check_draw():
    if '' not in board:
        return True
    return False

# Function to restart the game and display the homepage
def restart_game(_):
    global board, current_player, player1, player2
    board = [''] * 9
    current_player = 'X'
    player1 = ''
    player2 = ''
    winner_msg.value = ''
    player1_input.value = ''
    player2_input.value = ''
    for button in buttons:
        button.description = ''
        button.disabled = False
        button.style.button_color = None  # Reset button color
    display_homepage()

# Function to display the game interface
def display_game_interface():
    clear_output()
    display(game_page_layout)

# Function to display the homepage
def display_homepage():
    clear_output()
    display(homepage_layout)

# Create input widgets for player names
player1_input = widgets.Text(placeholder='Enter Player 1 Name')
player2_input = widgets.Text(placeholder='Enter Player 2 Name')
start_button = widgets.Button(description='Start Game')
start_button.on_click(start_game)
error_msg = widgets.Label(value='', style={'color': 'red'})

# Create homepage layout
homepage_layout = widgets.VBox([
    widgets.Label(value="Welcome to Tic-Tac-Toe!"),
    widgets.Label(value="Enter Player Names:"),
    player1_input,
    player2_input,
    start_button,
    error_msg
])

# Create game board buttons
buttons = [widgets.Button(description='', layout=widgets.Layout(width='80px', height='80px')) for _ in range(9)]
for i, button in enumerate(buttons):
    button.on_click(lambda _, i=i: handle_move(buttons[i], i))

# Create game interface
game_board = widgets.GridBox(buttons, layout=widgets.Layout(grid_template_columns='repeat(3, 80px)'))
restart_button = widgets.Button(description='Restart Game', button_style='warning')
restart_button.on_click(restart_game)
winner_msg = widgets.Label(value='')

# Create game page layout
game_page_layout = widgets.VBox([
    widgets.Label(value="Tic-Tac-Toe Game"),
    game_board,
    winner_msg,
    restart_button
])

# Display the homepage initially
display_homepage()
```
